```python
import requests
import warnings

# 1. Suppress the InsecureRequestWarning.
#    This warning is displayed when making an HTTPS request with `verify=False`.
#    While not recommended for production, it's acceptable for this assignment
#    since we are intentionally using a self-signed certificate.
from requests.packages.urllib3.exceptions import InsecureRequestWarning
warnings.simplefilter('ignore', InsecureRequestWarning)

# 2. Define the target server URL and the city to query.
#    The URL must use 'https' to communicate with our secure server.
city_to_check = "Tokyo"
url = f"https://127.0.0.1:5000/weather/{city_to_check}"

print(f"  Sending a request to the secure proxy server for '{city_to_check}'...")
print(f"   Target URL: {url}\n")

try:
    # 3. Send a GET request to the server.
    #    - verify=False: This is the most crucial part for this client.
    #      It tells the `requests` library to skip the SSL certificate verification.
    #      Without this, the request would fail because our certificate is self-signed
    #      and not trusted by a public Certificate Authority (CA).
    response = requests.get(url, verify=False)

    # 4. Print the response from the server.
    print(f"  Request successful!")
    print(f"   Response Status Code: {response.status_code}")

    if response.status_code == 200:
        print("\n   [Response Data (JSON)]")
        # The .json() method parses the JSON response into a Python dictionary.
        weather_data = response.json()
        print(f"   - City: {weather_data.get('name')}")
        print(f"   - Temperature: {weather_data.get('main', {}).get('temp')} °C")
        print(f"   - Weather: {weather_data.get('weather', [{}])[0].get('main')}")
    else:
        # If the server returns an error code, print the error message.
        print("\n   [Error Message]")
        print(f"   {response.text}")

except requests.exceptions.SSLError as e:
    print(f" An SSL error occurred: {e}")
except requests.exceptions.RequestException as e:
    # This block catches network errors, e.g., if the server is not running.
    print(f" A request error occurred. Make sure the server is running.")
    print(f"   Error details: {e}")
```
